```python
class FileManagement():
    
    def __init__(self):
        
        self._name = "records.txt"
        
        self._data = ""
        
        self._file = open(self._name, 'a+')
            
        
    def _isExist(self, filepath):
        return os.path.exists(filepath)
    
    def insert_data(self, data):
        
        self._file = open(self._name, 'a+')
        
        self._data = "\n".join(data)
        
        self._data += "\n\n"
        
        try:
            self._file.write(self._data)
            self._file.close()
            return True
        except Exception as e:
            return False
        return False
    
    def read_data(self):
        final_data = []
        try:
            total_chunk = open(self._name, 'r').read()
            
            total_data = total_chunk.split("\n\n")
            
            if len(total_data) == 0:
                return 5
            
            for item in total_data[:-1]:
                
                chunk = item.split("\n")
                
                final_data.append((chunk[0],chunk[1],chunk[2]))
                
            return final_data
        
        except:
            return False
        
    def delete_data(self, id):
        
        data = self.read_data()
        
        if data == 5:
            return False
        elif not data:
            return False
        else:
            del data[id]
        
        final_data = ""
        
        for elem in data:
            final_data += "\n".join(elem)
            final_data += "\n\n"
        try:
            self._file = open(self._name, 'w+')
            self._file.write(final_data)
            self._file.close()
            return True
        except Exception as e:
            return False
        return False
            
    def edit_data(self, id, new_data):
        data = self.read_data()
        
        if data == 5:
            return False
        elif not data:
            return False
        else:
            data[id] = (new_data[0], new_data[1], new_data[2])
            
        final_data = ""
        
        for elem in data:
            final_data += "\n".join(elem)
            final_data += "\n\n"
        try:
            self._file = open(self._name, 'w+')
            self._file.write(final_data)
            self._file.close()
            return True
        except Exception as e:
            return False
        return False
```

```python
def main():
    
    fm = FileManagement()
    
    print("\n1. Insert a record")
    print("2. View all record")
    print("3. Delete a record")
    print("4. Update a Record")
    
    choice = input("\nEnter your choice : ")
    
    if choice == "1":
        name = input("\nEnter a name : ")
        email = input("\nEnter a email : ")
        password = input("\nEnter password : ")
        
        if fm.insert_data([name, email, password]):
            print("\nData Inserted Successfully!")
        else:
            print("Oops! Something went wrong!")
            
    elif choice == "2":
        
        data = fm.read_data()
        
        if data == 5:
            print("\nNo records found!")
        elif not data:
            print("\nOops! Something went wrong!")
        else:
            
            print("\n:: All Records ::\n")
            
            for index, elem in enumerate(data):
                
                print("\nSl No :" + str(index+1))
                print("Name : " + elem[0])
                print("Email : " + elem[1])
                print("Password : " + elem[2])
                print()
    elif choice == "3":
        
        sl_no = input("\nEnter the Sl No of the record : ")
        
        if fm.delete_data(int(sl_no)-1):
            print("Record Deleted Successfully!")
        else:
            print("Oops! Something went wrong!")
            
    elif choice == "4":
        sl_no = input("\nEnter the Sl No of the record : ")
        
        name = input("\nEnter a name : ")
        email = input("\nEnter a email : ")
        password = input("\nEnter password : ")
        
        if fm.edit_data(int(sl_no)-1, [name, email, password]):
            print("Record Updated Successfully!")
        else:
            print("Oops! Something went wrong!")
            
    else:
        print("\nBad Choice! Dont mess up!")
```

```python
main()
```
